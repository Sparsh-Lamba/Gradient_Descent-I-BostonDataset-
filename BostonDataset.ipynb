```python
def gd(dataval,alpha,num_iteration):
    m=np.zeros(dataval.shape[1])
    for i in range(num_iteration):
        m=stepgrad(dataval,alpha,m)
    print("Cost=", cost(dataval,m))
    return m
```

```python
def stepgrad(dataval,alpha,m):
    N=dataval.shape[0]
    updated_m=np.zeros(dataval.shape[1])
    for i in range(N):
        x=dataval[i,:-1]
        x=np.append(x,1)
        y=dataval[i,-1]
        for j in range(dataval.shape[1]):
            updated_m[j] += (-2/N)*(y-(m*x).sum())*x[j]
        m=m-(alpha*updated_m[j])
    return m
```

```python
def cost(dataval, m):
    costval = 0
    N = dataval.shape[0]
    for i in range(N):
        x = dataval[i,:-1]
        x = np.append(x, 1)
        y = dataval[i,-1]
        costval +=(1/N)* ((y-(m*x).sum())**2)
        
    return float(costval)
```

```python
def run():
    alpha=0.0001
    num_iteration=1000
    M=gd(traindata,alpha,num_iteration)
    m=M[0:13]
    c=M[13]
    return m,c
```

```python
m,c=run()
print(m,c)
```

```python
def predict(dataval, m, c):
    N = dataval.shape[0]
    y = np.zeros(N)
    for i in range(N):                               #prediction using m ,c ,alpha and numof iterations 
        X = dataval[i,:]
        y[i] = (m * X).sum() + c
    return y
```

```python
testpred=predict(testdata,m,c)
testpred
```

```python
from sklearn import preprocessing
scale=preprocessing.StandardScaler()
val1=traindata
scale.fit(val1)
val1=scale.transform(val1)
scaledtrainy=val1[:,-1] # scaled training data output 
scaledtrainy
```

```python
testtrain=val1[:,:-1]
ans=predict(testtrain,m,c)
ans=ans.reshape(-1,1)           #prediction of training data using gradient descent and then scaling 
scale.fit(ans)
scaledtrainpred=scale.transform(ans)
scaledtrainpred.reshape(-1)
```

```python
def score(scaledtrainy,scaledpredy):                                    #comparing of actualtraining data output 
    u=((scaledtrainy-scaledpredy)**2).sum()                             #and output predicted of training data
    v=((scaledtrainy-scaledtrainy.mean())**2).sum()
    return 1- u/v  
```

```python
scaledpredy=scaledtrainpred.reshape(-1)
score(scaledtrainy,scaledpredy)
```
